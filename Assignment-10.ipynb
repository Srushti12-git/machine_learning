## Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

**Answer:**
Anomaly detection is the identification of rare items, events, or observations which raise suspicions by differing significantly from the majority of the data.

#### **1. Point Anomalies**

* A **Point Anomaly** occurs when an individual data instance is considered anomalous with respect to the rest of the data points in a global context. 

* This is the simplest and most common form of anomaly.

* **Detailed Explanation:** Point anomalies represent a single, isolated incident that is unusual. They are typically identified using statistical thresholds (such as Z-scores or Interquartile Range) or distance-based algorithms like K-Nearest Neighbors (KNN). In these cases, the point simply exists far outside the normal distribution of the rest of the population.

* **Example:** A credit card transaction of **$5,000** for a user whose historical spending habits consistently range between **$20 and $100** per transaction[cite: 10]. The transaction itself is flagged simply because its value is an extreme outlier compared to the rest of the user's data.

#### **2. Contextual Anomalies**
* A **Contextual Anomaly** (also known as a conditional anomaly) occurs when a data instance is considered anomalous only when viewed within a specific **context**. 

* If the same data point occurred in a different context, it would be considered perfectly normal.

* **Detailed Explanation:** To identify these, one must define two types of attributes:

    * **Contextual Attributes:** The circumstances (e.g., time, location, or season).

    * **Behavioral Attributes:** The actual value being measured (e.g., temperature, energy usage).

    This type is highly prevalent in time-series data where "normal" is a moving target.

* **Example:** A temperature reading of **35°C** is normal during a summer afternoon. However, that same **35°C** reading would be a significant anomaly if recorded in the middle of **winter** in a cold region. The anomaly is not the value itself, but the value given the time of year.

#### **3. Collective Anomalies**

A **Collective Anomaly** occurs when a collection or sequence of related data instances is anomalous relative to the entire dataset, even if the individual points within that set appear normal when viewed in isolation.

* **Detailed Explanation:** These occur when the structural relationship between a group of data points breaks the expected pattern or trend. The anomaly arises from the specific combination, order, or frequency of the points. Detecting these requires analyzing the connection between data points rather than evaluating them independently.

* **Example:** A single **$1 bank transfer** is a standard, non-suspicious event. However, a **rapid sequence of 500 transfers of $1** occurring within a 60-second window is a collective anomaly. The individual $1 values are "normal," but their collective frequency suggests a hacking attempt, "smurfing," or a denial-of-service attack.

## Question 3: What are the key components of a Time Series?

**Answer :**
A time series is a sequence of data points indexed in time order. To effectively analyze and forecast time series data, it is typically decomposed into four distinct components that represent the underlying patterns within the data.

The four key components are as follows:

#### **1. Trend (T)**

The **Trend** represents the long-term progression of the series. It reflects the underlying direction (upward, downward, or stationary) that the data is moving toward over an extended period.

* **Detailed Explanation:** Trends are often caused by structural changes such as population growth, technological shifts, or long-term economic shifts. While there may be short-term fluctuations, the trend focuses on the "big picture" trajectory.

* **Example:** **Global Population Growth.** Despite local variations or short-term mortality spikes, the long-term trend of the world population has been a consistent upward move over the last century.

#### **2. Seasonality (S)**

**Seasonality** refers to periodic fluctuations that repeat at regular, fixed intervals within a single year (such as daily, weekly, monthly, or quarterly).

* **Detailed Explanation:** Seasonal patterns are predictable and are usually tied to the calendar, the weather, or specific human traditions. Unlike cycles, the duration of a seasonal period is fixed and known.

* **Example:** **Spike in toy sales in December.** Retailers consistently see a massive increase in sales during the holiday season every year, which is a predictable seasonal event.

#### **3. Cyclicity (C)**

**Cyclicity** (or Cycle) represents long-term oscillations or fluctuations in the data that do not have a fixed period. 

* **Detailed Explanation:** Unlike seasonality, cycles typically last longer than a year and their duration can be highly variable. These are often linked to external factors like business cycles, political shifts, or economic climates.

* **Example:** **Economic Recessions.** The economy goes through periods of expansion and contraction (the "boom and bust" cycle). While these cycles repeat, they do not occur at the same time every year and can last anywhere from 2 to 10 years.

#### **4. Residual / Noise (I or E)**

The **Residual** (also known as Irregular or Noise) represents the random, unpredictable variations that remain after the trend, seasonality, and cyclicity have been accounted for.

* **Detailed Explanation:** This component is the result of unforeseen events, sudden shocks, or "noise" in the data collection process. It has no discernible pattern and cannot be easily modeled or predicted.

* **Example:** **A stock price dip due to an unexpected news event.** If a CEO suddenly resigns or a natural disaster occurs, the immediate and erratic change in a company’s stock price is considered noise because it was not part of a trend or a seasonal pattern.

## Question 4: Define Stationarity in time series.

**Answer:**
To select the right anomaly detection algorithm, it is essential to understand how each approach defines an "outlier." Below is a detailed comparison of three widely used methods:

#### **1. Isolation Forest**
* **Approach:** This algorithm is based on the principle of **Isolation**. It uses a collection of random decision trees (Random Forest) to partition data. Since anomalies are few and different, they are much easier to "isolate" from the rest of the data. In practice, anomalies will have significantly shorter average path lengths in the trees than normal points.
* **Suitable Use Cases:** * **High-Dimensional Data:** It performs exceptionally well when there are many features.
    * **Large Datasets:** It has a linear time complexity, making it very fast.
    * **Financial Fraud:** Detecting credit card transactions that deviate from a user's standard behavior.

#### **2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**
* **Approach:** DBSCAN is a **Density-Based** clustering algorithm. It groups together points that are closely packed (points with many nearby neighbors) and marks points that lie alone in low-density regions as "noise" or outliers.
* **Suitable Use Cases:**
    * **Spatial Data:** Excellent for identifying clusters of arbitrary shapes (like a crescent moon).
    * **Noise Removal:** Used when you need to separate clear clusters from background interference.
    * **Geographical Analysis:** Identifying clusters of crime or disease outbreaks in a city map.

#### **3. Local Outlier Factor (LOF)**
* **Approach:** LOF is a **Relative Density**-based method. Instead of looking at global density, it compares the local density of a point to the local densities of its neighbors. A point is considered an outlier if its local density is significantly lower than that of its neighbors.
* **Suitable Use Cases:**
    * **Varying Densities:** Ideal for datasets where "normal" density changes across different regions of the feature space.
    * **Local Anomalies:** Finding a point that looks normal globally but is an outlier compared to its immediate surroundings.
    * **Sensor Networks:** Detecting a malfunctioning sensor in a network where different geographic regions have naturally different baseline readings.

## Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX.

**Answer :**

| Model | Full Name | Core Structure & Logic | Key Components | Suitable Application |
| :--- | :--- | :--- | :--- | :--- |
| **AR** | **Autoregressive** | Predicts future values based on a linear combination of its own past (lagged) values. | **p** (Lag order) | Stationary data where the current state depends on previous states (e.g., daily temperature). |
| **MA** | **Moving Average** | Predicts future values based on past forecast errors (residuals) rather than the values themselves. | **q** (Error lag order) | Data influenced by short-term, random "shocks" that eventually return to the mean. |
| **ARIMA** | **AutoRegressive Integrated Moving Average** | Combines AR and MA models and includes an "Integration" step to handle non-stationary data. | **(p, d, q)** where **d** is differencing | Non-seasonal datasets that exhibit a clear trend but no repeating cycles. |
| **SARIMA** | **Seasonal ARIMA** | An extension of ARIMA that explicitly supports univariate time series data with a seasonal component.  | **(p, d, q) × (P, D, Q)s** where **s** is the period | Data with repeating patterns, such as monthly airline passenger counts or summer electricity demand. |
| **SARIMAX** | **Seasonal ARIMA with Exogenous Variables** | The most robust model; it includes all SARIMA features plus the ability to factor in external variables. | **SARIMA + X** (Exogenous variables) | Complex forecasting where external factors (e.g., weather or price changes) impact the target variable. |

## Question 7: Isolation Forest on NYC Taxi Data

```python
from sklearn.ensemble import IsolationForest
import seaborn as sns

# Load sample data
url = 'https://raw.githubusercontent.com/numenta/NAB/master/data/realKnownCause/nyc_taxi.csv'
df_taxi = pd.read_csv(url, parse_dates=['timestamp']).head(500)

# Isolation Forest
iso = IsolationForest(contamination=0.05)
df_taxi['anomaly'] = iso.fit_predict(df_taxi[['value']])

# Visualize
plt.figure(figsize=(10,6))
sns.scatterplot(x=df_taxi.index, y=df_taxi['value'], hue=df_taxi['anomaly'], palette='coolwarm')
plt.title('Anomalies in NYC Taxi Fare')
plt.show()
```

## Question 8: SARIMA Forecast

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit Model
model = SARIMAX(df['Passengers'], order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit()

# Forecast
forecast = results.get_forecast(steps=12)
plt.plot(df, label='Actual')
plt.plot(forecast.predicted_mean, label='Forecast', color='red')
plt.legend()
plt.show()
```

## Question 9: Local Outlier Factor (LOF)

```python
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

# Generate Data
X = np.random.randn(100, 2)
X = np.r_[X, np.random.uniform(low=-4, high=4, size=(10, 2))]

# Apply LOF
lof = LocalOutlierFactor(n_neighbors=20)
y_pred = lof.fit_predict(X)

# Plot
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap='Paired')
plt.title('LOF Anomaly Detection')
plt.show()
```

## Question 10: Real-time Power Grid Workflow

**Answer:**
Working as a data scientist for a power grid monitoring company requires a robust, scalable workflow to handle streaming data every 15 minutes. Below is the explained workflow for forecasting energy demand and detecting abnormal consumption patterns.

#### **1. Anomaly Detection (Isolation Forest)**

For streaming data, I would choose **Isolation Forest**.

* **Reasoning:** Isolation Forest is highly efficient for high-frequency streaming data because it has a linear time complexity and a small memory footprint. Unlike DBSCAN or LOF, it does not require a computationally expensive distance or density calculation between all points, making it ideal for real-time 15-minute intervals.

* **Application:** It would be used to identify "Point Anomalies" such as sudden spikes (potential equipment failure or surges) or drops (power outages).

#### **2. Short-Term Forecasting (SARIMAX)**

For forecasting energy usage, I would use the **SARIMAX** model.

* **Reasoning:** Energy consumption is highly seasonal (daily and weekly cycles), requiring the "Seasonal" component of SARIMA[cite: 52]. Furthermore, the dataset includes **weather conditions**, which are external factors that directly impact energy usage.

* **Application:** SARIMAX allows us to include weather as an **Exogenous (X)** variable to significantly improve the accuracy of short-term demand forecasts.

#### **3. Validation and Monitoring**

* **Validation Strategy:** I would use a **Time-Series Walk-Forward Validation**. Instead of a standard train-test split, the model is trained on a sliding window of historical data and tested on the next 15-minute observation to mimic real-world production.

* **Metrics:** 
    * **Forecasting:** Mean Absolute Percentage Error (MAPE) to track demand accuracy.

    * **Anomalies:** Precision and Recall to ensure we aren't overwhelming operators with false alarms.

* **Monitoring:** Implement "Drift Detection" to alert the team if the model's performance degrades due to changing weather patterns or consumer behavior shift.

#### **4. Business Impact and Operations**

* **Resource Allocation:** Accurate forecasting helps the business decide how much power to generate or purchase in advance, reducing waste.

* **Preventative Maintenance:** Real-time anomaly detection can alert engineers to localized grid stress before it leads to a total blackout.

* **Operational Efficiency:** Automated monitoring reduces the need for manual oversight of thousands of grid sensors.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 1. Simulate Power Grid Data (15-min intervals)
data = {
    'timestamp': pd.date_range(start='2026-01-01', periods=100, freq='15min'),
    'usage': np.random.normal(100, 10, 100).tolist(),
    'temp': np.random.normal(25, 5, 100).tolist()
}
df = pd.DataFrame(data).set_index('timestamp')

# Inject a simulated anomaly (spike)
df.iloc[95, 0] = 500 

# 2. Real-time Anomaly Detection (Isolation Forest)
iso_model = IsolationForest(contamination=0.02, random_state=42)
df['anomaly_status'] = iso_model.fit_predict(df[['usage']]) 
# -1 indicates an anomaly

# 3. Short-term Forecasting (SARIMAX)
# Using 'temp' as an exogenous variable
model = SARIMAX(df['usage'], exog=df['temp'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 4))
results = model.fit(disp=False)

# Forecast the next 15 minutes (requires a weather forecast for 'temp')
next_temp_forecast = [[24]] 
forecast = results.get_forecast(steps=1, exog=next_temp_forecast)

print("--- Real-time Grid Monitoring Report ---")
print(f"Latest Usage: {df['usage'].iloc[-1]:.2f} kW")
print(f"Anomaly Detected: {'YES' if df['anomaly_status'].iloc[-1] == -1 else 'NO'}")
print(f"Next 15-min Demand Forecast: {forecast.predicted_mean.values[0]:.2f} kW")
```
