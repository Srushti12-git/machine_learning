Question 1: What is Simple Linear Regression?
Answer:Simple Linear Regression is a statistical method used to model the relationship between two continuous variables: one Independent Variable ($X$) and one Dependent Variable ($Y$).

The goal is to determine if a change in $X$ results in a predictable change in $Y$, allowing us to make forecasts or understand the strength of their connection.
1. The Mathematical Equation

The model is represented by the equation of a straight line:

$$Y = \beta_0 + \beta_1X + \epsilon$$

Where:

$Y$: The Dependent Variable (Target/Response).

$X$: The Independent Variable (Predictor/Feature).

$\beta_0$ (Beta Zero): The Intercept. This is the value of $Y$ when $X$ is zero.

$\beta_1$ (Beta One): The Slope. This represents the change in $Y$ for every 1-unit increase in $X$.

$\epsilon$ (Epsilon): The Error Term. This accounts for the variation in $Y$ that the line cannot explain (the "noise").

2. The Goal: Line of Best Fit

The algorithm uses a method called Ordinary Least Squares (OLS). It calculates the line that minimizes the sum of the squares of the vertical distances (residuals) between each data point and the line itself.

Think of it as trying to draw a line through a cloud of dots so that the total "distance" from all dots to the line is as small as possible.

3. Real-World Example: Play Store Analytics

In our current analysis of the Google Play Store dataset, we can use Simple Linear Regression to answer specific questions:

The Question: Can we predict the number of Installs ($Y$) based solely on the number of Reviews ($X$)?

The Result: The model will give us a slope ($\beta_1$). If $\beta_1$ is $0.5$, it suggests that for every new review received, the app is likely to gain roughly $0.5$ additional installs.

4. Key Assumptions

For the results of a Simple Linear Regression to be valid, a few conditions must be met:

=>Linearity: The relationship between $X$ and $Y$ must be roughly a straight line (not a curve).

=>Independence: Each observation must be independent of the others.

=>Homoscedasticity: The "spread" of the errors should be consistent across all values of $X$.

=>Normality: The error terms ($\epsilon$) should follow a normal distribution.

Question 3: What is heteroscedasticity, and why is it important to address in regression models?
Answer:Heteroscedasticity:-
Heteroscedasticity occurs when the variance of the residuals (the "errors" or the distance between the data points and the regression line) is not constant across all levels of an independent variable.

In a perfect model (Homoscedastic), the "noise" or uncertainty is the same everywhere. In a heteroscedastic model, the uncertainty grows or shrinks as the input values change.
If you ignore heteroscedasticity, your regression model (OLS) suffers from three major flaws:

A. Unreliable Standard Errors

While the model's coefficients ($\beta_1$) might still be roughly correct, the Standard Errors will be biased. Since standard errors are used to calculate p-values, you might conclude that a feature is "statistically significant" when it actually isn't (or vice versa).

B. Inaccurate Confidence Intervals

Because the error is not uniform, the "margin of error" the model provides will be too narrow in some areas and too wide in others. This makes your predictions untrustworthy for high-value or low-value inputs.

C.Ordinary Least Squares(OLS) is no longer "BLUE"

In statistics, OLS is ideally the Best Linear Unbiased Estimator (BLUE). Under heteroscedasticity, OLS is no longer the most efficient way to estimate the relationship because it treats all data points as equally "certain," even though the "noisy" points should technically carry less weight.

Question 4: What is Multiple Linear Regression?
Answer:
Multiple Linear Regression is an extension of Simple Linear Regression used to predict the outcome of a continuous dependent variable ($y$) based on the values of two or more independent variables ($x_1, x_2, \dots, x_n$).
While Simple Linear Regression asks, "How does $X$ affect $Y$?", Multiple Regression asks, "How do $X_1, X_2,$ and $X_3$ together affect $Y$?"
1. The Mathematical Formula
The relationship is expressed as a multi-dimensional linear equation:


$$y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n + \epsilon$$

Where:
->$y$: The Dependent Variable (Target).

->$x_1, x_2, \dots$: The Independent Variables (Features/Predictors).

->$\beta_0$: The Intercept (The value of $y$ when all $x$ are zero).

->$\beta_1, \beta_2, \dots$: The Partial Coefficients. They represent the change in $y$ for a 1-unit change in that specific $x$, holding all other variables constant.

->$\epsilon$: The Error term (Residuals).

2. Why Use Multiple Regression?

In the real world, single factors rarely explain an outcome.
Case Study: Your "Calories Burnt" Project

 We are predicting energy expenditure. A simple model might only look at Duration. However, a Multiple Regression model looks at:

$x_1$: Duration

$x_2$: Heart Rate

$x_3$: Body Temperature

$x_4$: Age

Result: By combining these features, we achieved an $R^2$ of 0.97, which is much higher than what any single feature could provide alone.

3. The Multicollinearity Challenge

In MLR, we face a problem not found in simple regression: Multicollinearity.

Definition: This occurs when independent variables are highly correlated with each other (e.g., trying to use both Height in Inches and Height in Centimeters to predict weight).

The Risk: It makes it difficult for the model to determine the individual effect of each feature, causing the coefficients ($\beta$) to become unstable and unreliable.

The Fix: We use the Variance Inflation Factor (VIF) to identify and remove redundant features.

4. Evaluation: $R^2$ vs. Adjusted $R^2$

As you add more variables to a model, the standard $R^2$ will always stay the same or increase, even if you add "garbage" data (like the color of the user's shirt to predict calories).

Adjusted $R^2$: This metric penalizes you for adding features that do not improve the model's predictive power.

Question 5: What is polynomial regression, and how does it differ from linear regression?

Understanding Polynomial Regression

Polynomial Regression is a form of regression analysis in which the relationship between the independent variable $x$ and the dependent variable $y$ is modeled as an $n^{th}$ degree polynomial.

While it fits a non-linear curve to the data, it is still considered a form of Linear Regression because the model is linear in terms of its unknown parameters (the coefficients $\beta$).

1. The Mathematical Equation

A simple linear regression follows $y = \beta_0 + \beta_1x$. Polynomial regression adds power terms to this foundation:

$$y = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + \dots + \beta_nx^n + \epsilon$$

Where:

->$n$: The Degree of the polynomial.

->$x^2, x^3, \dots$: These are the original features raised to a power, allowing the model to "bend" the line to fit curves.

2. Key Differences: Linear vs. Polynomial
| Feature | Linear Regression | Polynomial Regression |
| :---- | :---- | :---- |
| **Visual Shape** | A straight line. | A curved line (parabola, S-curve, etc.). |
| **Relationship** | Assumes ![][image2] changes at a constant rate with ![][image1]. | Captures relationships where the rate of change is not constant. |
| **Complexity** | Simple, low risk of overfitting. | Complex, high risk of overfitting as degree increases. |
| **Use Case** | Height vs. Weight. | Engine Efficiency vs. Temperature. |


3. The "Degree" and Overfitting

Choosing the right degree ($n$) is the most critical part of Polynomial Regression:

1.Underfitting (Degree 1): Using a straight line for data that is clearly curved. The model is too simple.

2.Optimal Fit (Degree 2 or 3): The curve follows the general trend of the data points without being distracted by individual "noise."

3.Overfitting (High Degree, e.g., 10+): The curve wiggles excessively to touch every single data point. While the error on training data is zero, the model will fail completely on new, unseen data.

Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.
(Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=[1,2,3,4,5]
Y=[2.1,4.3,6.1,7.9,10.2]

#convert the list to dataframe
df=pd.DataFrame({'X':X,'Y':Y})

#train test split
X=df[['X']]
Y=df[['Y']]


```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()

model.fit(X,Y)

print(f"Slope (Beta 1): {model.coef_}")
print(f"Intercept (Beta 0): {model.intercept_}")
print("Predictions for X [1, 2, 3, 4, 5]:")
Y_pred=model.predict(X)
print(Y_pred)

from sklearn.metrics import r2_score

slope = model.coef_.item()
intercept = model.intercept_.item()
r2 = r2_score(Y, Y_pred)
```

```python
plt.figure(figsize=(10, 6))

# Plot the original data points
plt.scatter(X, Y, color='blue', label='Actual Data Points', s=100)

# Plot the regression line
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line (Best Fit)')

# Aesthetics
plt.title('Simple Linear Regression: X vs Y', fontsize=14)
plt.xlabel('Independent Variable (X)', fontsize=12)
plt.ylabel('Dependent Variable (Y)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Display the equation on the plot
# These f-strings will now work because 'slope' and 'intercept' are pure numbers
plt.text(1, 9, f'Equation: Y = {slope:.2f}X + {intercept:.2f}', color='darkred', fontsize=12, fontweight='bold')
plt.text(1, 8.3, f'R² Score: {r2:.4f}', color='black', fontsize=11)
```

Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results. (Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

df=pd.DataFrame({'Area':Area,'Rooms':Rooms,'Price':Price})

X=df[['Area','Rooms']]
Y=df[['Price']]
```

```python
from sklearn.linear_model import LinearRegression

multi_model=LinearRegression()
multi_model.fit(X,Y)
print("Model parameters")
print(f'Slope is {multi_model.coef_}')
print(f'Intercept is {multi_model.intercept_}')
Y_pred=multi_model.predict(X)
print('The predicted values of Y are:')
print(Y_pred)
```

```python
plt.scatter(Y,Y_pred)
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
#Predicting r2 sore
print(f'The R2 score is {r2_score(Y,Y_pred)}')
print(f'The mean absolute error is {mean_absolute_error(Y,Y_pred)}')
```

```python
#Checking variance inflation factor for all features
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Features']=df.columns
vif['VIF']=[variance_inflation_factor(df.values,i) for i in range(len(df.columns))]
print(vif)
#VIF > 10: High Multicollinearity.VIF for Area is highest so must drop Area
```

Question 8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5] 
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X=np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])
```

```python
from sklearn.preprocessing import PolynomialFeatures

#Transform Features to Polynomial (Degree 2)

poly_features=PolynomialFeatures(degree=2,include_bias=False)
X_train=poly_features.fit_transform(X)
X_train
```

```python
from sklearn.linear_model import LinearRegression

#Fit the Linear Regression model to the transformed features

poly_model=LinearRegression()
poly_model.fit(X_train,Y)
```

```python
#Generate points for a smooth curve plot
#We create 100 points between 1 and 5 to show the "bend" in the curve

X_range=np.linspace(1,5,100).reshape(-1,1)
X_range_poly=poly_features.transform(X_range)
Y_range_pred=poly_model.predict(X_range_poly)
```

```python
from sklearn.metrics import r2_score

#Predictions for original X points to calculate R2
Y_pred=poly_model.predict(X_train)
r2=r2_score(Y,Y_pred)
```

```python
#Visualisation
plt.figure(figsize=(10,6))
plt.scatter(X,Y,color='black',label='Actual Data',s=100)
plt.plot(X_range,Y_range_pred,color='red',linewidth=3,label='Polynomial Fit (Deg 2)')
# Aesthetics
plt.title('Polynomial Regression: Degree 2 Fit', fontsize=14)
plt.xlabel('Independent Variable (X)', fontsize=12)
plt.ylabel('Dependent Variable (Y)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Display metrics on plot
plt.text(1, 13, f'R² Score: {r2:.4f}', fontsize=12, fontweight='bold', color='darkgreen')

plt.show()

# --- Console Output Summary ---
print("--- Polynomial Model Results ---")
print(f"Intercept: {poly_model.intercept_:.4f}")
print(f"Coefficients (1, X, X^2): {poly_model.coef_}")
print(f"R-squared Score: {r2:.4f}")
```

Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.
(Include your Python code and output in the code box below.)
Answer:
Question

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.array([10,20,30,40,50]).reshape(-1,1)
Y=np.array([15,35,40,50,65])
```

```python
#Fit the Linear Regression model to the transformed features

from sklearn.linear_model import LinearRegression
resi_model=LinearRegression()
resi_model.fit(X,Y)

Y_pred=resi_model.predict(X)
residuals = Y - Y_pred
```

```python
#Visualization: Residuals Plot
plt.figure(figsize=(10, 6))

# Plotting the Residuals against the input X
plt.scatter(X, residuals, color='purple', s=100, edgecolor='black', label='Residuals')

# Add a horizontal line at zero (The reference line for perfect predictions)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Zero Error Line')

# Aesthetics (Building on the styles in the Canvas)
plt.title('Residuals Plot: Checking for Heteroscedasticity', fontsize=14, fontweight='bold')
plt.xlabel('Independent Variable (X)', fontsize=12)
plt.ylabel('Residuals (Errors)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Adding a limit to the y-axis to see the spread clearly
plt.ylim(min(residuals)-5,max(residuals)+5)

plt.show()
```

```python
print("--- Model Predictions & Residuals ---")
for i,val in enumerate(X.flatten()):
    print(f'X:{val} | Actual Y:{Y[i]} Predicted Y:{Y_pred[i]:.2f} | Residuals : {residuals[i]:.2f}')
print("\n" + "="*40)
print("--- HETEROSCEDASTICITY ASSESSMENT ---")
print("="*40)
```

Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

# **Strategy Report: Addressing Heteroscedasticity and Multicollinearity in Real Estate Pricing Models**

In real estate modeling, features like **Area** and **Number of Rooms** are often highly correlated, and **House Prices** typically exhibit increasing variance as the price increases. To ensure the model provides stable coefficients and reliable confidence intervals, I would implement the following two-phased approach.

## **Phase 1: Mitigating Multicollinearity**

Multicollinearity makes individual feature weights  unstable. If Area and Rooms move in perfect sync, the model doesn't know which one is actually driving the price.

1. **VIF Analysis:** I would calculate the **Variance Inflation Factor (VIF)** for all features. Any feature with a VIF score above 5 or 10 indicates a redundancy problem.  
2. **Feature Selection:** If Area and Rooms are redundant, I would likely drop "Rooms" because "Area" (square footage) is usually a more granular and powerful predictor of market value.  
3. **Feature Engineering:** Alternatively, I could create an **interaction term** or a **ratio**, such as "Average Room Size" (Area / Rooms), which captures the relationship between the two without the mathematical redundancy.  
4. **Regularization (Ridge/Lasso):** I would implement **Ridge Regression**. The $ L2 $ penalty shrinks the coefficients of correlated features, distributing the weights more evenly and preventing any single redundant feature from having an erratic coefficient.

## **Phase 2: Correcting Heteroscedasticity**

Heteroscedasticity (the "fan shape" in residuals) means our model is less certain about expensive homes than cheap ones. This invalidates standard p-values.

1. **Target Transformation (Log-Scaling):** I would apply a **Natural Log Transformation** to the house price (![][image2]). Since real estate prices are often log-normally distributed, this "squashes" the high-end variance, making the spread of errors constant across the entire price range.  
2. **Weighted Least Squares (WLS):** Instead of treating every house as equally important, I would use WLS to assign smaller weights to high-variance observations (e.g., luxury zip codes) and higher weights to more stable data points.  
3. **Robust Standard Errors:** If the variance persists after transformations, I would utilize **Huber-White (HC3) Robust Standard Errors**. This does not change the line itself, but it corrects the standard errors and p-values so that our statistical significance tests remain valid.

## **Summary for Stakeholders**

By applying these steps, we move from a model that is "accurate but fragile" to one that is **robust and interpretable**. This ensures that when we tell a client that "an extra 100 sq. ft. adds ₹5,00,000 to the value," that number is based on a stable statistical foundation, not on a mathematical fluke caused by redundant data or skewed variance.
