Question 1: What is K-Nearest Neighbors (KNN) and how does it work?

Answer:
K-Nearest Neighbors (KNN) is a non-parametric, supervised learning algorithm used for both classification and regression tasks. It is often referred to as a "lazy learner" because it does not learn a discriminative function from the training data but instead stores the dataset and performs computation during the prediction phase.

# How KNN Works

The algorithm operates on the fundamental principle of feature similarity—the idea that similar data points exist in close proximity to each other.

* Distance Calculation: When a new, unlabeled data point is introduced, the algorithm calculates the distance between that point and every other point in the training dataset.

* Metric Selection: While Euclidean distance is the most common metric used to measure this "closeness," other metrics like Manhattan or Minkowski distance can also be applied.

* Identifying Neighbors: The algorithm selects the $K$ number of training data points that are closest to the new point.

* Making a Prediction:
    * In Classification: The algorithm uses "majority voting." The new data point is assigned to the class that is most frequent among its $K$ nearest neighbors.

    * In Regression: Instead of a vote, the algorithm predicts the output by taking the average (mean) or the median of the values of the $K$ nearest neighbors.
    

# The Role of 'K'

The choice of the value $K$ is a critical hyperparameter:

* Small $K$ (e.g., $K=1$): The model is very sensitive to noise in the data, which can lead to overfitting.

* Large $K$: The model becomes more stable but may overlook smaller patterns, potentially leading to underfitting.

Question 2: What is the Curse of Dimensionality and how does it affect KNN? 

Answer:
The Curse of Dimensionality refers to various phenomena that arise when analyzing and organizing data in high-dimensional spaces (often hundreds or thousands of features) that do not occur in low-dimensional settings like 2D or 3D space.As the number of dimensions increases, the volume of the space grows exponentially, which has several negative impacts on the K-Nearest Neighbors (KNN) algorithm:

1. Data Sparsity

In high-dimensional space, the data points become very sparse. To maintain the same level of density as in lower dimensions, you would need an exponential increase in the amount of training data. Since we usually have a fixed number of samples, the "neighbors" found by KNN end up being very far away from the query point.

2. Distance Dilution

KNN relies on distance metrics (like Euclidean distance) to find similarity.In high dimensions, the distance between any two points in the dataset tends to converge to the same value.

* The difference between the distance to the nearest neighbor and the distance to the farthest neighbor becomes negligible.

* Because all points seem almost equally far away, the concept of a "nearest neighbor" becomes mathematically meaningless.

3. Overfitting and Noise

With too many dimensions, the KNN model is more likely to pick up on random noise or irrelevant features rather than the actual underlying patterns. This leads to poor generalization on new, unseen data.

4. Computational Complexity

As dimensions grow, the computational cost of calculating distances between the query point and every training sample increases significantly. This makes KNN—which is already a "lazy learner"—extremely slow to provide predictions in real-time environments.To combat these issues, data scientists often use dimensionality reduction techniques like PCA (Principal Component Analysis) before applying KNN to ensure the distances remain meaningful and the model remains efficient.

Question 3: What is PCA and how is it different from feature selection?

Answer:
Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms a high-dimensional dataset into a smaller set of variables while retaining as much of the original information (variance) as possible. It achieves this by creating new, uncorrelated variables called Principal Components.

# How PCA Works

PCA identifies the directions (eigenvectors) in which the data varies the most. By projecting the data onto these directions, PCA allows us to summarize the data using fewer dimensions without losing the critical patterns that define the dataset.

### Comparison: Feature Selection vs. PCA (Feature Extraction)

| Feature | Feature Selection | PCA (Feature Extraction) |
| :--- | :--- | :--- |
| **Nature of Features** | Selects a subset of the **original** features.  | Creates **entirely new** features (Principal Components).  |
| **Transformation** | **No transformation**; features keep their original meaning (e.g., "Age", "Income").  | **Linear transformation**; original features are combined, making them harder to interpret.  |
| **Data Retention** | **Discards** the unselected features entirely.  | Information from **all** original features is compressed into the new components.  |
| **Use Case** | Best when model **interpretability** is a high priority.  | Best for **reducing noise** and overcoming the "Curse of Dimensionality". [cite: 11, 16] |

# Why use PCA over Feature Selection?

In complex datasets, like the Wine dataset or gene expression data, individual features may contain a lot of noise. PCA is often preferred in these cases because it filters out that noise by focusing on the components with the highest eigenvalues, which represent the most significant signals in the data.

Question 4: What are eigenvalues and eigenvectors in PCA? 

Answer:
In the context of Principal Component Analysis (PCA), eigenvalues and eigenvectors are the mathematical "building blocks" that allow us to transform a high-dimensional dataset into its most informative components. They are derived from the covariance matrix of the data.

1. Eigenvectors: The Directions
Eigenvectors represent the directions of the new feature space.

* When you run PCA, the first eigenvector points in the direction where the data is most spread out (the maximum variance).

* Each subsequent eigenvector is perpendicular (orthogonal) to the ones before it, capturing the next highest level of variation.

* These eigenvectors essentially become the Principal Components themselves.

2. Eigenvalues: The Magnitudes
Eigenvalues represent the magnitude or the "strength" of their corresponding eigenvectors.

* An eigenvalue tells us how much variance is captured in the direction of its eigenvector.

* A large eigenvalue means that the specific Principal Component explains a significant portion of the data's total information.

* A small eigenvalue suggests that the component captures very little information and might be representing noise.

# Why are they important?
Eigenvalues and eigenvectors are critical for three main reasons:

* Ranking Components: We use eigenvalues to rank the Principal Components from most important to least important.

* Dimensionality Reduction: By looking at the eigenvalues, we can decide to keep only the top $k$ components that account for the majority of the variance (e.g., 95%) and discard the rest.

* Data Compression: They allow us to project complex, multi-dimensional data onto a simpler 2D or 3D coordinate system while preserving the "shape" and relationships of the original data.

Question 5: How do KNN and PCA complement each other? 

Answer:
When applied together in a single machine learning pipeline, KNN and PCA act as complementary forces that solve each other's inherent limitations. While KNN is a simple and effective classifier, it struggles with high-dimensional data—a problem that PCA is perfectly designed to fix.Here is how they complement each other:

1. Overcoming the "Curse of Dimensionality"

KNN relies on calculating distances (like Euclidean distance) to find neighbors. In high-dimensional spaces, these distances become diluted, and all points start to look equally far away. PCA reduces the number of dimensions by projecting the data onto its most informative principal components. This ensures that the distances calculated by KNN remain meaningful and accurate.

2. Automatic Noise Reduction

Raw datasets often contain "noise"—features that don't help with classification and only confuse the model. During the PCA process, components with very low eigenvalues (which usually represent noise) are discarded. When KNN is trained on this "cleaned" PCA data, it performs better because it is focusing only on the most significant biological or statistical signals.

3. Improving Computational Speed

KNN is a "lazy learner," meaning it has to calculate the distance to every single training point every time it makes a prediction. If your dataset has 1,000 features, this is computationally expensive. By using PCA to reduce those 1,000 features to perhaps 20 principal components, you drastically reduce the number of calculations KNN has to perform, making the model much faster in production environments

4. Handling Feature Correlation

KNN assumes that features are independent, but in real-world data (like gene expression), many features are highly correlated. PCA creates new, orthogonal (uncorrelated) components. Providing these uncorrelated components to KNN prevents the model from giving "double importance" to redundant information.

Question 6: Train a KNN Classifier on the Wine dataset with and without feature
scaling. Compare model accuracy in both cases.
(Include your Python code and output in the code box below.)


```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load Wine dataset [cite: 27]
wine = load_wine()
X, y = wine.data, wine.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. KNN without scaling
knn_unscaled = KNeighborsClassifier(n_neighbors=5)
knn_unscaled.fit(X_train, y_train)
acc_unscaled = accuracy_score(y_test, knn_unscaled.predict(X_test))

# 2. KNN with scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
acc_scaled = accuracy_score(y_test, knn_scaled.predict(X_test_scaled))

print(f"Accuracy without scaling: {acc_unscaled:.4f}")
print(f"Accuracy with scaling: {acc_scaled:.4f}")
```

Comparison & Analysis

* Without Scaling: The accuracy is significantly lower. This happens because features with large magnitudes dominate the Euclidean distance calculation, effectively making the model "blind" to smaller but equally important features.

* With Scaling: By using StandardScaler, all features are shifted to have a mean of 0 and a standard deviation of 1. This gives every feature an equal opportunity to influence the classification, leading to a much higher and more reliable accuracy.

Question 7: Train a PCA model on the Wine dataset and print the explained variance
ratio of each principal component.
(Include your Python code and output in the code box below.)

```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# 1. Load the dataset
wine = load_wine()
X = wine.data

# 2. Standardize the features (Essential for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Initialize and fit PCA
# We don't specify n_components so we can see all of them
pca = PCA()
pca.fit(X_scaled)

# 4. Extract and print the explained variance ratio
explained_variance = pca.explained_variance_ratio_

print("Explained Variance Ratio per Principal Component:\n")
for i, ratio in enumerate(explained_variance):
    print(f"PC{i+1}: {ratio:.4f} ({ratio*100:.2f}%)")

# Calculate cumulative variance
cumulative_variance = explained_variance.cumsum()
print(f"\nTotal variance explained by first 2 components: {cumulative_variance[1]*100:.2f}%")
```

Question 8: Train a KNN Classifier on the PCA-transformed dataset (retain top 2
components). Compare the accuracy with the original dataset.
(Include your Python code and output in the code box below.)

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

# 1. Load the Wine dataset
wine = load_wine()
X, y = wine.data, wine.target

# 2. Split into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Scale the data (Mandatory for both KNN and PCA)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- Case 1: KNN on Original (Scaled) Dataset ---
knn_original = KNeighborsClassifier(n_neighbors=5)
knn_original.fit(X_train_scaled, y_train)
acc_original = accuracy_score(y_test, knn_original.predict(X_test_scaled))

# --- Case 2: KNN on PCA-Transformed Dataset (Top 2 Components) ---
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_train_pca, y_train)
acc_pca = accuracy_score(y_test, knn_pca.predict(X_test_pca))

# Output Results
print(f"Accuracy on Original Dataset (13 features): {acc_original:.4f}")
print(f"Accuracy on PCA Dataset (2 features):      {acc_pca:.4f}")
```

Comparison & Analysis 

* Original Dataset Accuracy: This baseline uses all 13 features. While high, it includes all the "noise" and potentially redundant information present in the raw features.

* PCA Dataset Accuracy: Even with only 2 features (components), the accuracy remains very high—and sometimes even improves. This is because PCA filters out less significant variance (noise) that might have confused the KNN algorithm in a higher-dimensional space.

* Efficiency: The PCA model is much faster to compute because KNN only needs to calculate distances in a 2D space rather than a 13D space.

Question 9: Train a KNN Classifier with different distance metrics (euclidean,
manhattan) on the scaled Wine dataset and compare the results.
(Include your Python code and output in the code box below.)

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# 1. Load and scale the dataset
wine = load_wine()
X, y = wine.data, wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 2. Compare Distance Metrics
metrics = ['euclidean', 'manhattan']
results = {}

for m in metrics:
    knn = KNeighborsClassifier(n_neighbors=5, metric=m)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    results[m] = acc
    print(f"Accuracy with {m.capitalize()} distance: {acc:.4f}")
```

Analysis

* Performance: In many high-dimensional datasets like the Wine dataset, Manhattan distance can sometimes outperform Euclidean distance. This is because Manhattan distance is less sensitive to outliers in individual dimensions compared to the squared terms used in the Euclidean formula.

* Metric Selection: While Euclidean is the default, switching to Manhattan is a common hyperparameter tuning step to see if the data "clusters" more effectively under a different geometric interpretation.

1. Using PCA to Reduce Dimensionality

Since gene expression data often contains thousands of features (genes) for only a small number of patients, PCA is used to transform these features into a set of linearly uncorrelated variables called Principal Components (PCs).

* Standardization: Before applying PCA, I would scale the data using StandardScaler so that genes with naturally higher expression levels do not dominate the variance calculation.

* Feature Extraction: PCA will condense the thousands of original genes into a few dozen components that capture the maximum variance (information) in the dataset.

2. Deciding How Many Components to Keep

Retaining the right number of components is critical to avoiding the Curse of Dimensionality.

* Explained Variance Ratio: I would analyze the explained_variance_ratio_ for each component.

* The 95% Rule: A common standard is to keep enough components to account for 95% of the total variance, ensuring most information is preserved while discarding noise.

* Scree Plot: I would use a Scree Plot to find the "elbow"—the point where adding more components provides diminishing returns in explained variance.

3. Using KNN for Classification Post-Reduction

Once the data is transformed into its principal components, the KNN Classifier is applied to the reduced dataset.

* Similarity Search: KNN will find the $K$ closest patients in the reduced PCA space and classify the new patient based on the majority label of those neighbors.

* Efficiency: Because PCA has reduced the number of dimensions, KNN can calculate distances much faster and more accurately than it could on the raw data.

4. Evaluating the Model

To ensure the model is reliable for medical diagnostics, I would use a multi-metric evaluation:

* Stratified K-Fold Cross-Validation: This ensures that each "fold" of the evaluation has a representative mix of each cancer type, which is vital if some cancers in the dataset are rarer than others.

* Confusion Matrix: This helps identify where the model is misclassifying (e.g., confusing Cancer Type A for Type B).F1-Score: Since accuracy can be misleading in medical data, the F1-Score provides a balanced view of Precision and Recall.

5. Stakeholder Justification

I would justify this pipeline to stakeholders as a robust biomedical solution using these key points:

* Prevents Overfitting: By reducing thousands of features down to the most informative components, we prevent the model from "memorizing" specific patients and instead force it to learn general patterns.

* Noise Filtration: Gene expression data is notoriously "noisy." PCA acts as a filter, keeping the significant biological signals and discarding random fluctuations.

* Computational Speed: This pipeline is efficient enough to provide near-instant classifications as new patient data becomes available.
